## Questão 1

Atualize o algoritmo para determinar se um número inteiro e positivo é par ou ímpar, usando uma laço condicional para aceitar apenas números maiores ou iguais a zero.

#### Pseudocódigo
```java
ALGORTIMO verifica_par_impar
DECLARE num, resto: INTEIRO

INICIO

    num = -1    
    ENQUANTO num < 0 FAÇA
        ESCREVA "Digite um número maior ou igual a zero:"
        LEIA num
    FIM_ENQUANTO
	resto ← num % 2
    SE resto == 0 ENTAO
        ESCREVA "O número é par!"
    SENAO
        ESCREVA "O número é impar!"
    FIM_SE

FIM
```

### Implementação em Python

````python
def verifica_par_impar():
    num = -1
    while num < 0:
        num = int(input("Digite um número maior ou igual a zero: "))

    resto = num % 2
    if resto == 0:
        print("O número é par!")
    else:
        print("O número é ímpar!")

if __name__ == "__main__":
    verifica_par_impar()
````

## Questão 2

Faça um algoritmo que exiba na tela uma contagem de 0 até n, exibindo apenas os múltiplos de 3.

#### Pseudocódigo
```java
ALGORTIMO MultiploTres
DECLARE n: INTEIRO

INICIO

	ESCREVA "Digite a quantidade de números:"
	LEIA n
	PARA i DE 0 ATÉ n+1 PASSO 3 FAÇA
		ESCREVA i
	FIM_PARA

FIM
```

### Implementação em Python

````python
def multiplo_tres():
    n = int(input("Digite a quantidade de números: "))
    for i in range(0, n + 1, 3):
        print(i, end=" ")

if __name__ == "__main__":
    multiplo_tres()
````

## Questão 3

Escreva um programa que leia a nota de diversos alunos, até que seja digitada uma nota negativa.
Nesse momento, ele mostra a média aritmética de todas as notas lidas e quantas notas foram lidas.
Ex. Foram lidas 14 notas. A média aritmética é 6.75!

### Pseudocódigo
```java
ALGORTIMO QuantMedia
DECLARE nota, soma, media: REAL; cont: INTEIRO

INICIO

	ESCREVA "Digite a nota do aluno (nota negativa finaliza): "
	LEIA nota
	soma <- 0
	cont <- 0
	ENQUANTO nota >= 0 FAÇA
		soma <- soma + nota
		cont <- cont + 1
		LEIA nota
	FIM_ENQUANTO
	SE cont > 0 ENTÃO
		media <- soma / cont
		ESCREVA "Foram lidas", cont, "nota(s). A média aritmética é", media
	FIM_SE

FIM
```

### Implementação em Python

````python
def quant_media():
    print("Digite a nota do aluno (nota negativa finaliza): ")
    nota = float(input())
    soma = 0
    cont = 0

    while nota >= 0:
        soma += nota
        cont += 1
        nota = float(input())

    if cont > 0:
        media = soma / cont
        print(f"Foram lidas {cont} nota(s). A média aritmética é {media:.1f}")

if __name__ == "__main__":
    quant_media()
````

## Questão 4

Dada uma sequência de números inteiros, calcular a sua soma.
Por exemplo, para a sequência {12, 17, 4, -6, 8, 0}, o seu programa deve escrever o número 35.

### Pseudocódigo

```java
ALGORITMO SomaValores
DECLARE n,i: INTEIRO; soma,num: REAL

INICIO

	ESCREVA "Digite a quantidade de números:"
	LEIA n
	soma <- 0
	i <- 1
	ENQUANTO i <= n FAÇA
		ESCREVA "Digite o número", i,":"
		LEIA num
		soma <- soma + num
		i <- i + 1
	FIM_ENQUANTO
	ESCREVA "A soma dos número é", soma

FIM
```

### Implementação em Python

````python
def soma_valores():
    n = int(input("Digite a quantidade de números: "))
    soma = 0
    i = 1

    while i <= n:
        num = float(input(f"Digite o número {i}: "))
        soma += num
        i += 1

    print("A soma dos números é", soma)

if __name__ == "__main__":
    soma_valores()
````
